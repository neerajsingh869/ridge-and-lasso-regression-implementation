```python
# Loading necessary dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# To ignore the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('df_train.csv')
dataset.head()
```

```python
dataset.shape
```

### Train-Test split

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(dataset, test_size=0.3, random_state=42)
```

```python
train_set
```

```python
test_set
```

```python
y_train = train_set["SalePrice"]
X_train = train_set.drop("SalePrice", axis = 1)
```

```python
y_test = test_set["SalePrice"]
X_test = test_set.drop("SalePrice", axis = 1)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lr = LinearRegression()
mse = cross_val_score(lr, X_train, y_train, cv = 5, scoring='neg_mean_squared_error')
mean_mse = np.mean(mse)
print(mean_mse)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-2, 5, 10, 20, 40, 70, 80, 100]}
ridge_regression = GridSearchCV(ridge, parameters, scoring = 'neg_mean_squared_error', cv = 5)
ridge_regression.fit(X_train, y_train)
```

```python
print(ridge_regression.best_params_)
print(ridge_regression.best_score_)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-2, 5, 10, 20, 40, 70, 80, 100]}
lasso_regression = GridSearchCV(lasso, parameters, scoring = 'neg_mean_squared_error', cv = 5)
lasso_regression.fit(X_train, y_train)
```

```python
print(lasso_regression.best_params_)
print(lasso_regression.best_score_)
```

### Predictions

```python
prediction_ridge = ridge_regression.predict(X_test)
prediction_lasso = lasso_regression.predict(X_test)
```

```python
sns.distplot(y_test-prediction_ridge)
```

```python
sns.distplot(y_test-prediction_lasso)
```
